```python
import matplotlib.pyplot as plt

# Data to visualize
models = ['DT', 'GB', 'KNN', 'LR', 'RF', 'SVM']
accuracy_percentage = [64.16, 64.16, 64.16, 68.34, 64.16, 69.70]

# Create a figure and axis
fig, ax = plt.subplots(figsize=(8, 5))

# Create a bar chart
bars = ax.bar(models, accuracy_percentage, color='skyblue')

# Add grid lines in the background
ax.yaxis.grid(True, linestyle='--', which='both', color='gray', alpha=0.7)

# Add titles and labels
ax.set_title('Model Accuracy Comparison (Percentage)')
ax.set_xlabel('Models')
ax.set_ylabel('Accuracy (%)')
ax.set_ylim(60, 75)

# Show the accuracy values on top of the bars in percentage format
for bar, v in zip(bars, accuracy_percentage):
    ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.2, f"{v:.2f}%", ha='center', fontsize=12)

# Display the plot
plt.show()
```

```python
import numpy as np

# Convert the DataFrame to a format suitable for grouped bar chart
df_grouped = df_percentage_transposed

# Identify unique accuracy values across models and years
unique_values = np.unique(df_grouped.values)

# Create a color map for the unique values
colors = plt.cm.tab20(np.linspace(0, 1, len(unique_values)))

# Plotting the grouped bar chart
fig, ax = plt.subplots(figsize=(10, 6))

width = 0.1  # Width of the bars

# Loop over each year to plot bars for each model
for i, year in enumerate(df_grouped.columns):
    bar_positions = np.arange(len(df_grouped.index)) + i * width
    ax.bar(bar_positions, df_grouped[year], width=width, label=f'Year {year}', color=[colors[np.where(unique_values == val)[0][0]] for val in df_grouped[year]])

# Add labels, title, and legend
ax.set_xlabel('Models')
ax.set_ylabel('Accuracy (%)')
ax.set_title('Model Accuracy Over Different Years (Percentage)')
ax.set_xticks(np.arange(len(df_grouped.index)) + width * (len(df_grouped.columns) / 2 - 0.5))
ax.set_xticklabels(df_grouped.index)
ax.legend(title='Year')
ax.grid(True, linestyle='--', which='both', color='gray', alpha=0.7)

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
models = ['RF', 'DT', 'GB', 'LR', 'SVC', 'KNN']
split = [76.53, 72.48, 75.77, 82.02, 71.45, 63.41]
gridsearch = [75.89, 71.94, 74.55, 73.99, 0.00, 62.55]
cv = [88.52, 86.68, 88.64, 77.55, 70.49, 71.80]

# Bar width
bar_width = 0.25

# Set position of bar on X axis
r1 = np.arange(len(models))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

# Colors
split_color = '#4CAF50'  # Soft green
gridsearch_color = '#FFC107'  # Warm yellow
cv_color = '#2196F3'  # Soft blue

# Make the plot
plt.figure(figsize=(10, 6))
plt.bar(r1, split, color=split_color, width=bar_width, edgecolor='grey', label='Split')
plt.bar(r2, gridsearch, color=gridsearch_color, width=bar_width, edgecolor='grey', label='GridSearch')
plt.bar(r3, cv, color=cv_color, width=bar_width, edgecolor='grey', label='CV')

# Add labels and title
plt.xlabel('Models', fontweight='bold')
plt.ylabel('Accuracy (%)', fontweight='bold')
plt.title('Model Performance Across Different Techniques')
plt.xticks([r + bar_width for r in range(len(models))], models)

# Add legend
plt.legend()

# Show the plot
plt.show()
```
